```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('customer_purchases.csv')
```

```python
print("Data Summary:")
print(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Annual_Income'], bins=20)
plt.title('Histogram of Annual Income')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Purchase_Frequency', y='Spending_Score', data=df)
plt.title('Purchase Frequency vs. Spending Score')
plt.show()
```

```python
corr_matrix = df[['Annual_Income', 'Spending_Score', 'Purchase_Frequency']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
print("Correlation Matrix:")
print(corr_matrix)
```

```python
print("Outliers in Spending Score:")
print(df['Spending_Score'].describe())
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(df['Spending_Score'])
plt.title('Boxplot of Spending Score')
plt.show()
```
