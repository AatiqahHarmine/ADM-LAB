```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('employee_performance.csv')
```

```python
print("Statistical Summaries:")
print(df.describe())
df = pd.get_dummies(df, columns=['Department'])
df['Experience_Level'] = pd.cut(df['Years_of_Experience'], bins=[0, 2, 5, 10], labels=['Junior', 'Mid-Level', 'Senior'])
grouped_df = df.groupby('Experience_Level')['Performance_Score'].mean().reset_index()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Monthly_Salary'], bins=20)
plt.title('Histogram of Monthly Salaries')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Years_of_Experience', y='Performance_Score', data=df)
plt.title('Years of Experience vs. Performance Score')
plt.show()
```

```python
corr_matrix = df[['Monthly_Salary', 'Years_of_Experience', 'Performance_Score']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Numerical Features')
plt.show()
```
